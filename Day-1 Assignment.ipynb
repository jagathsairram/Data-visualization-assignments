```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

# Question 1 

```python
a=np.arange(40,50)
b=np.arange(50,60)
plt.title("line plot a and b ",size=25)
plt.xlabel("X-axis",size=18)
plt.ylabel("Y-axis",size=18)
plt.plot(a,b,"rD-",linewidth=2,markersize=12)
```

# question 2

```python
days = [1,2,3,4,5,6,7] 
sales_1 = [160,150,140,145,175,165,180] 
sales_2 = [70,90,160,150,140,145,175] 
plt.title("Days vs Sales",size=25)
plt.xlabel("Days",size=17)
plt.ylabel("sales",size=17)
plt.plot(days,sales_1,"b+-",linewidth=2,markersize=10,label="sales of comapany 1")
plt.plot(days,sales_2,"r^-",linewidth=2,markersize=10,label="sales of company 2")
plt.legend(loc="lower right")
```

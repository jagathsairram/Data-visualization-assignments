```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Question-1

```python
df=sns.load_dataset('fmri')
df.head()
```

# Relplot

```python
sns.relplot(x='timepoint',y='signal',data=df,hue='event',style='region')
```

# Line plot

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=df,hue='event',style='region',row='region',col='event',row_order=['frontal','parietal'],markers=True)
```

# Box plot

```python
sns.catplot(x='event',y='signal',kind='box',data=df,hue='region')
```
